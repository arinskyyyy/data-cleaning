```python
from google.colab import drive, userdata
import os

# 1. Clean up the accidental Git setup in the wrong folder
!rm -rf /content/.git

# 2. Mount Drive and navigate to your ACTUAL folder
drive.mount('/content/drive')
%cd /content/drive/MyDrive/data-cleaning

# 3. Load your token and set your details
token = userdata.get('GITHUB_TOKEN')
username = "arinskyyyy"
repo = "data-cleaning"
remote_url = f"https://{token}@github.com/{username}/{repo}.git"

# 4. Configure Git and initialize in the CORRECT folder
!git config --global user.email "colab@example.com"
!git config --global user.name "Colab User"
!git init
!git branch -M main

# 5. Connect to GitHub
!git remote remove origin 2>/dev/null
!git remote add origin {remote_url}

# 6. Pull existing files from GitHub first
print("Pulling existing files from GitHub...")
!git config pull.rebase false
!git pull origin main --allow-unrelated-histories --no-edit

# 7. Add your actual folder contents
!git add .

# 8. PROMPT FOR COMMIT MESSAGE
# The script will pause here and wait for you to type in the output box below
print("\n" + "="*50)
commit_message = input("Enter your commit message for this update: ")
print("="*50 + "\n")

# 9. Commit with the user's message and push
!git commit -m "{commit_message}"
print("Pushing to GitHub...")
!git push -u origin main
```
